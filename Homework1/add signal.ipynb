```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt;

# 
Fs=100 # 
Size=1000 #
t=np.arange(0,Size)/Fs

plt.figure()

# 
plt.figure(1)
x1 = np.sin(2*np.pi*1*t)
plt.plot(t,x1)
plt.xlabel('t')
plt.ylabel('x1')
plt.title('$sin(t)$')
plt.show()

# 
plt.figure(2)
x2 = signal.square(2*np.pi*t)
plt.plot(t,x2)
plt.xlabel('t')
plt.ylabel('x2')
plt.title('square wave')
plt.show()

# 
plt.figure(3)
x3 = np.array(t>0, dtype=np.int64)
plt.plot(t,x3)
plt.xlabel('t')
plt.ylabel('x3')
plt.title('step signal')
plt.show()

# 
plt.figure(4)
y = x1+x2+x3
plt.plot(t,y)
plt.xlabel('t')
plt.ylabel('y')
plt.title('add')
plt.show()
```
